```python
# color hist
def extract_color_histogram(image, bins=(8,8,8)):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    hist = cv2.calcHist([hsv],[0,1,2], None, bins,
                       [0,180,0,256,0,256])
    if imutils.is_cv2():
        hist = cv2.normalize(hist)
    else:
        cv2.normalize(hist,hist)
    return hist.flatten()
```

Some functions for NN are below.

```python
def read_image(file_path):
    img = cv2.imread(file_path, cv2.IMREAD_COLOR) #cv2.IMREAD_GRAYSCALE
    return cv2.resize(img, (ROWS, COLS), interpolation=cv2.INTER_CUBIC)


def prep_data(images):
    count = len(images)
    data = np.ndarray((count, CHANNELS, ROWS, COLS), dtype=np.uint8)

    for i, image_file in enumerate(images):
        image = read_image(image_file)
        data[i] = image.T
        if i%250 == 0: print('Processed {} of {}'.format(i, count))
    
    return data
```

###### Now CV - We know that a single K accuracy is ~52% (varies with each run), which is not great. What happens if we do multiple fits with CV, and grid search for various K vaues?

```python
# let's grid search across our range of values
neighbors = 2
cv_scores = []

for k in range(1,neighbors):
    print("[INFO] Fitting with {} Neighbors".format(k))
    start = datetime.datetime.now()
    print(start)
    knn = KNeighborsClassifier(n_neighbors=neighbors,
                             n_jobs=-1)
    scores = cross_val_score(knn, features_train, labels_train, cv=5, scoring='accuracy')
    cv_scores.append(scores.mean())
    print("[INFO] histogram accuracy {}%".format(cv_scores))
    end = datetime.datetime.now()
    print(end)
    tdelta = end - start
    print("[INFO] time duration {}".format(divmod(tdelta.total_seconds(), 60)))
```

Now let's repeat the process with more K values. This could take some time!

```python
# let's grid search across our range of values
neighbors = [1,2,3,4,5,10,15,30,50]
cv_scores = []

for k in range(0,len(neighbors)):
    print("[INFO] Fitting with {} Neighbors".format(neighbors[k]))
    start = datetime.datetime.now()
    print(start)
    knn = KNeighborsClassifier(n_neighbors=neighbors[k],
                             n_jobs=-1)
    scores = cross_val_score(knn, features_train, labels_train, cv=5, scoring='accuracy')
    cv_scores.append(scores.mean())
    print("[INFO] histogram accuracy {}%".format(cv_scores[k]))
    end = datetime.datetime.now()
    print(end)
    tdelta = end-start
    print("[INFO] time duration {}".format(divmod(tdelta.total_seconds(), 60)))
```

```python
labels = []
for i in train_images:
    if 'dog' in i:
        labels.append(1)
    else:
        labels.append(0)

sns.countplot(labels)
sns.plt.title('Cats and Dogs')
```

```python
optimizer = RMSprop(lr=1e-4)
objective = 'binary_crossentropy'

def dognet():
    
    model = Sequential()

    model.add(Convolution2D(32, 3, 3, border_mode='same', input_shape=(3, ROWS, COLS), activation='relu'))
    model.add(Convolution2D(32, 3, 3, border_mode='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2), dim_ordering="th"))

    model.add(Convolution2D(64, 3, 3, border_mode='same', activation='relu'))
    model.add(Convolution2D(64, 3, 3, border_mode='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2), dim_ordering="th"))    

    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))
    
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))

    model.add(Dense(1))
    model.add(Activation('sigmoid'))

    model.compile(loss=objective, optimizer=optimizer, metrics=['accuracy'])
    return model


model = dognet()
```

We've set up a pretty shallow CNN with just a couple convolutional layers and a pooling step. Let's run this on just 3 epochs to start.

```python
nb_epoch = 3
batch_size = 16

## Callback for loss logging per epoch
class LossHistory(Callback):
    def on_train_begin(self, logs={}):
        self.losses = []
        self.val_losses = []
        
    def on_epoch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))
        self.val_losses.append(logs.get('val_loss'))

early_stopping = EarlyStopping(monitor='val_loss', patience=3, verbose=1, mode='auto')        
        
def run_dognet():
    history = LossHistory()

    start = datetime.datetime.now()
    print("[INFO] Starting NN fit on {}".format(start))
    model.fit(train, labels, batch_size=batch_size, epochs=nb_epoch,
              validation_split=0.25, verbose=0, shuffle=True, callbacks=[history, early_stopping])

    predictions = model.predict(test, verbose=0)
    end = datetime.datetime.now()
    print("[INFO] Finished fit on {}".format(end))
    
    tdelta = end-start
    print("[INFO] time duration {}".format(divmod(tdelta.total_seconds(), 60)))

    return predictions, history

predictions, history = run_dognet()
```

```python
loss = history.losses
val_loss = history.val_losses

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Trend')
plt.plot(loss, 'blue', label='Training Loss')
plt.plot(val_loss, 'green', label='Validation Loss')
plt.xticks(range(0,nb_epoch)[0::2])
plt.legend()
plt.show()
```

###### NN1 Summary

We have a poor (KNN level) performing nn with a shallow (4 layers deep) CNN. 

Let's now expand upon that and increase the complexity.

```python
optimizer = RMSprop(lr=1e-4)
objective = 'binary_crossentropy'

def dognet_2():
    
    model_16 = Sequential()

    model_16.add(Convolution2D(32, 3, 3, border_mode='same', input_shape=(3, ROWS, COLS), activation='relu'))
    model_16.add(Convolution2D(32, 3, 3, border_mode='same', activation='relu'))
    model_16.add(MaxPooling2D(pool_size=(2, 2), dim_ordering="th"))

    model_16.add(Convolution2D(64, 3, 3, border_mode='same', activation='relu'))
    model_16.add(Convolution2D(64, 3, 3, border_mode='same', activation='relu'))
    model_16.add(MaxPooling2D(pool_size=(2, 2), dim_ordering="th"))    

    model_16.add(Convolution2D(128, 3, 3, border_mode='same', activation='relu'))
    model_16.add(Convolution2D(128, 3, 3, border_mode='same', activation='relu'))
    model_16.add(MaxPooling2D(pool_size=(2, 2), dim_ordering="th"))    
    
    model_16.add(Convolution2D(256, 3, 3, border_mode='same', activation='relu'))
    model_16.add(Convolution2D(256, 3, 3, border_mode='same', activation='relu'))
    model_16.add(MaxPooling2D(pool_size=(2, 2), dim_ordering="th"))    

    model_16.add(Flatten())
    model_16.add(Dense(256, activation='relu'))
    model_16.add(Dropout(0.5))
    
    model_16.add(Dense(256, activation='relu'))
    model_16.add(Dropout(0.5))

    model_16.add(Dense(1))
    model_16.add(Activation('sigmoid'))

    model_16.compile(loss=objective, optimizer=optimizer, metrics=['accuracy'])
    return model_16


model_16 = dognet_2()
```

Let's run 3 epochs with 16 layers.

```python
SVG(model_to_dot(model_16).create(prog='dot', format='svg'))
```

```python
nb_epoch = 10
batch_size = 16

## Callback for loss logging per epoch
class LossHistory(Callback):
    def on_train_begin(self, logs={}):
        self.losses = []
        self.val_losses = []
        
    def on_epoch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))
        self.val_losses.append(logs.get('val_loss'))

early_stopping = EarlyStopping(monitor='val_loss', patience=3, verbose=1, mode='auto')        

def run_dognet_16():
    history = LossHistory()

    start = datetime.datetime.now()
    print("[INFO] Starting NN fit on {}".format(start))
    model_16.fit(train_nn, labels_train, batch_size=batch_size, epochs=nb_epoch,
              validation_split=0.25, verbose=0, shuffle=True, callbacks=[history, early_stopping])

    predictions = model_16.predict(test_nn, verbose=0)
    end = datetime.datetime.now()
    print("[INFO] Finished fit on {}".format(end))
    
    tdelta = end-start
    print("[INFO] time duration {}".format(divmod(tdelta.total_seconds(), 60)))

    return predictions, history

predictions, history = run_dognet_16()
```
